# CUSTOMER GENDER DISTRIBUTION

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender', palette='pastel', edgecolor='black')

plt.title('Customer Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig('gender_distribution')
plt.show()
```

# Gender distribution by percentage

```python
gender_counts = df['Gender'].value_counts()
gender_percent = df['Gender'].value_counts(normalize=True) * 100

print("Gender Distribution (Counts):")
print(gender_counts)
print("\nGender Distribution (Percentages):")
print(gender_percent.round(2))

# Plot
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender', palette='pastel', edgecolor='black')

plt.title('Customer Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
for i, count in enumerate(gender_counts):
    pct = gender_percent[i]
    plt.text(i, count+20, f"{pct:.1f}%", ha='center', fontsize=11, fontweight='bold')
    plt.savefig("gender_distribution_by_percentage")

plt.show()
```

```python
# Plot average purchase by category
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Category', y='Purchase Amount (USD)', 
            estimator='mean', palette='muted', edgecolor='black')

plt.title('Average Purchase Amount by Category')
plt.xlabel('Product Category')
plt.ylabel('Average Purchase Amount (USD)')
plt.savefig('average_purchase_amount_category')
plt.show()
```

# Payment method usage

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x='Payment Method', palette='Set2', edgecolor='black')

plt.title('Payment Method Usage')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')

# Rotate labels for readability
plt.xticks(rotation=30)

# Add counts on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.savefig('payment_method_usage')

plt.show()
```

# Machine learning

# Removing columns that will not help prediction

```python
df = df.drop([
    'Unnamed:0',
    'Customer ID',
    'Item Purchased',
    'Location'],axis=1,errors='ignore')
```

```python
print(df.columns)
```

```python
df = df.drop(
    columns = [
        'Unnamed:0',
        'Customer ID',
        'Item Purchased',
        'Location'
    ],
    errors='ignore'
)
print(df.head())
```

# Converting text columns to numbers (encoding)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
    if df [col].dtypes=='object':
        df[col]=le.fit_transform(
            df[col].astype(str)
        )

print(df.head())
```

# Creating x and y

```python
x = df.drop('Purchase Amount (USD)',axis=1)
y = df['Purchase Amount (USD)']
x
```

```python
y
```

# Spliting dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

# Model training

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=42)
model.fit(x_train,y_train)
```
